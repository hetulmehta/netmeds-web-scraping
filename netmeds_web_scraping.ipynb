```python
from bs4 import BeautifulSoup as bs 
import requests
import pandas as pd
from concurrent.futures import ThreadPoolExecutor
import time
```

```python
HEADERS = ({'Accept-Language': 'en-US, en;q=0.5'})
url='https://www.netmeds.com/prescriptions'
webpage =requests.get(url, headers=HEADERS)
soup = bs(webpage.content, "lxml")
```

```python
url=[]
ul=soup.find_all('ul',{"class":"alpha-drug-list"})
for i in range(0,len(ul)):
    li=ul[i].find_all('li')
    for j in range(0,len(li)):
        a=li[j].find('a')
        url.append(a.get("href"))
```

```python
len(url)
```

```python
links=[]
for i in url:
    webpage =requests.get(i, headers=HEADERS)
    soup = bs(webpage.content, "lxml")
    items=soup.find_all('li',{"class":"product-item"})
    print(i,len(items))
    for j in range(0,len(items)):
        anchor=items[j].find('a')
        links.append(anchor.get("href"))
```

```python
len(links)
```

```python
def info(i):
    webpage =requests.get(i, headers=HEADERS)
    soup = bs(webpage.content, "lxml")
    
    try:
        name=soup.find('h1',{"class":"black-txt"}).text
    except:
        name=""
    try:
        category=soup.find('span',{"class":"gen_drug ellipsis"}).text
    except:
        category=""
    try: 
        descrip={}
        druginfo = soup.find_all(lambda tag: tag.name == 'div' and tag.get('class') == ['druginfo-dv'])
        if len(druginfo)>0:
            for drug in range(0,len(druginfo)):
                h2key=druginfo[drug].find("h2")
                lival=druginfo[drug].find("li")
                if(h2key):
                    descrip[h2key.text]=lival.text
            tab1=soup.find("div",{"id":"np_tab1"})
            tab2=soup.find("div",{"id":"np_tab2"})
            tab3=soup.find("div",{"id":"np_tab3"})
            tab4=soup.find("div",{"id":"np_tab4"})
            tab5=soup.find("div",{"id":"np_tab5"})
            tab6=soup.find("div",{"id":"np_tab6"})
            h2=tab1.find("h2").text
            li=tab1.find("li").text
            descrip[h2]=li
            doc=tab2.find('div',{"id":'war-talk-to-your-doctor'})
            p=tab2.find("p").text
            li=doc.find_all('li')
            l1=[]
            dic={}
            for l in range(0,len(li)):
                l1.append(li[l].text)
            ltext=" ; ".join(l1)
            dic[p]=ltext
            doc=tab2.find('div',{"id":'war-pregnancy-and-breast-feeding'})
            li=doc.find_all('li')
            l1=[]
            for l in range(0,len(li)):
                l1.append(li[l].text)
            ltext=" ; ".join(l1)
            dic['Pregnancy & Breast Feeding']=ltext
            doc=tab2.find('div',{"id":'war-driving-and-using-machines'})
            p=tab2.find("p").text
            li=doc.find_all('li')
            l1=[]
            for l in range(0,len(li)):
                l1.append(li[l].text)
            ltext=" ; ".join(l1)
            dic['driving and using machines']=ltext
            doc=tab2.find('div',{'id':'war-children-and-adolescents'})
            p=tab2.find("p").text
            li=doc.find_all('li')
            l1=[]
            for l in range(0,len(li)):
                l1.append(li[l].text)
            ltext=" ; ".join(l1)
            dic['children and adolescents']=ltext
            descrip['Warning & Precautions']=dic
            dic={}
            p=tab3.find("p").text
            li=tab3.find_all('li')
            l1=[]
            for l in range(0,len(li)):
                l1.append(li[l].text)
            ltext=" ; ".join(l1)
            dic[p]=ltext
            descrip['Interactions']=dic
            li=tab4.find_all('li')
            l1=[]
            for l in range(0,len(li)):
                l1.append(li[l].text)
            ltext=" ; ".join(l1)
            descrip['Directions & For Use']=ltext
            
            doc=tab5.find('div',{"id":'serious-side-effects'})
            li=doc.find_all('li')
            l1=[]
            dic={}
            for l in range(0,len(li)):
                l1.append(li[l].text)
            ltext=" ; ".join(l1)
            dic['Serious']=ltext
            doc=tab5.find('div',{"id":'common-side-effects'})
            li=doc.find_all('li')
            l1=[]
            for l in range(0,len(li)):
                l1.append(li[l].text)
            ltext=" ; ".join(l1)
            dic['Common']=ltext
            doc=tab5.find('div',{"id":'uncommon-side-effects'})
            li=doc.find_all('li')
            l1=[]
            for l in range(0,len(li)):
                l1.append(li[l].text)
            ltext=" ; ".join(l1)
            dic['Uncommon']=ltext
            doc=tab5.find('div',{"id":'very-rare-side-effects'})
            li=doc.find_all('li')
            l1=[]
            for l in range(0,len(li)):
                l1.append(li[l].text)
            ltext=" ; ".join(l1)
            dic['Very Rare']=ltext
            descrip['Side Effects']=dic
            li=tab6.find_all('li')
            l1=[]
            for l in range(0,len(li)):
                l1.append(li[l].text)
            ltext=" ; ".join(l1)
            descrip['More Information']=ltext
            
        else:
            des=soup.find_all('div',{'class':"inner-content"})
            for n in range(0,len(des)):
                h2=des[n].find('h2').text
                p=des[n].find_all('p')
                p1=[]
                for para in range(0,len(p)):
                    p1.append(p[para].text)
                ptext=" ".join(p1)
                descrip[h2]=ptext
                if(len(p1)==0):
                    li=des[n].find_all('li')
                    l1=[]
                    for l in range(0,len(li)):
                        l1.append(li[l].text)
                    ltext=" ; ".join(l1)
                    descrip[h2]=ltext
    except Exception as e:
        descrip=""

    try:
        det={}
        d=soup.find('div',{"id":"hideshow"})
        ul=d.find_all('ul')
        for ix in range(0,len(ul)):
            div=ul[ix].find_all('div')
            key=div[1].text
            value=div[2].text
            det[key]=value
        details=det   
    except:
        details=""
    try:
        img=[]
        pp=soup.find_all('figure',{"class":"figure largeimage"})
        print(len(pp))
        for x in range(0,len(pp)):
            p=pp[x].find("img")
            input_tag=p.get("src")
            img.append(input_tag) 
        im=' ; '.join(img)
        img_url=im 
    except:
        img_url=""
    try:
        sp=soup.find('span',{"class":"final-price"}).text
        sp=sp.replace("*","")
    except:
        sp=""
    try:
        mrp=soup.find('span',{"class":"price"}).text
    except:
        mrp=""
        
    return name,category,descrip,details,img_url,sp,mrp,i
if __name__ == '__main__':
    with ThreadPoolExecutor(max_workers = 8) as executor:
        results = executor.map(info,links)
    res=[]
    for i in results:
        res.append(i)  
    name=[]
    cat=[]
    des=[]
    det=[]
    img=[]
    spec=[]
    sp=[]
    mrp=[]
    produrl=[]
    for i in range(0,len(res)):
        name.append(res[i][0])
        cat.append(res[i][1])
        des.append(res[i][2])
        det.append(res[i][3])
        img.append(res[i][4])
        sp.append(res[i][5])
        mrp.append(res[i][6])
        produrl.append(res[i][7])
    df=pd.DataFrame({'product':name,'category':cat,'selling price':sp,'market price':mrp,'description':des,
                 'details':det,'images':img,'producturl':produrl})
```

```python
import re
import numpy as np
df1 = df.replace(np.nan, '', regex=True)
def Clean_names(City_name):
  if(re.search('\d', City_name)):
      pos=re.search('\d', City_name).start()
      return City_name[pos:]
  else:
    return City_name
          
# Updated the city columns
df1['market price'] = df1['market price'].apply(Clean_names)
df1['selling price'] = df1['selling price'].apply(Clean_names)
```
